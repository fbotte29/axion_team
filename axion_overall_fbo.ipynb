```python
#supprimer des colonnes
df = df.drop(['visits_lc', 'orders_lc','demand_sales_lc_(usd)'], axis=1)
```

```python
# remplissage des colonnes numérique avec des 0
#df['audience_strategy'].fillna(0, inplace=True)
cols_a_remplacer = ['audience_details','amount_spent','impressions','reach','clicks','videoviews','video_views_100%','traffic','orders','web_sales','new_customer','orders_retargeting','store_sales','video_25%','video_50%','video_75%','engagement']
df[cols_a_remplacer] = df[cols_a_remplacer].fillna(0)
```

```python
# Calcul du pourcentage total de valeurs nulles par colonne
null_percentage = round((df.isnull().sum() / len(df) * 100),2)
high_null_columns = null_percentage[null_percentage > 30]
print(f"{high_null_columns}")
```

```python
df.info()
```

```python
# dataframe pour analyse statistique
df_stat = df[['campaign_type','audience_strategy','format','platform','amount_spent','impressions','reach','clicks','videoviews','video_views_100%','traffic','orders','web_sales','new_customer','orders_retargeting','store_sales','video_25%','video_50%','video_75%','engagement']]
```

```python
df_stat.info()
```

```python
# Appliquer le One-Hot Encoding aux colonnes `object` catégorielle
df_encoded = pd.get_dummies(df_stat, columns=['campaign_type','audience_strategy','format','platform'],drop_first=True)
```

```python
matrix_correlation = df_encoded.corr()
print(matrix_correlation)
```

```python
import plotly.express as px

fig = px.imshow(
    matrix_correlation,
    #text_auto='.2f',       # Afficher les valeurs (2 décimales)
    color_continuous_scale="aggrnyl",  # Palette de couleurs
    aspect="auto",         # Ajuster automatiquement les proportions
)

# Personnalisation
fig.update_layout(
    title="Matrice de Corrélation",
    title_font_size=20,    # Taille du titre
    xaxis_title="Variables",
    yaxis_title="Variables",
    coloraxis_colorbar=dict(
        title="Corrélation"
    ),
    width=800,             # Largeur de la figure
    height=800             # Hauteur de la figure
)
fig.show()
```

```python
# Filtrer les corrélations > 0.6
import numpy as np
filtered_matrix = matrix_correlation.where(matrix_correlation > 0.6, other=np.nan)
```

```python
fig = fig = px.imshow(
    filtered_matrix,
    #text_auto='.2f',       # Afficher les valeurs (2 décimales)
    color_continuous_scale="aggrnyl",  # Palette de couleurs
    aspect="auto",         # Ajuster automatiquement les proportions
)

# Personnalisation
fig.update_layout(
    title="Matrice de Corrélation",
    title_font_size=20,    # Taille du titre
    xaxis_title="Variables",
    yaxis_title="Variables",
    coloraxis_colorbar=dict(
        title="Corrélation"
    ),
    width=800,             # Largeur de la figure
    height=800             # Hauteur de la figure
)
fig.show()
```

```python
# calcul CPC  somme amount_spent / somme Clicks
cpc = round(df['amount_spent'].sum() / df['clicks'].sum(),2)
```

```python
# calcul CTR  somme Clicks / somme impressions
ctr = round(df['clicks'].sum() / df['impressions'].sum(),2)
```

```python
#datafram cpc
df_kpi = df[['amount_spent','clicks','impressions']]
```

```python
import numpy as np

# Calcul du CPC par ligne
df_kpi['CPC'] = np.where(df['clicks'] > 0, df['amount_spent'] / df['clicks'], 0)
df_kpi
```

```python
# Calcul du CTR par ligne
df_kpi['CTR'] = np.where(df['clicks'] > 0, df['clicks'] / df['impressions'],0)*100
df_kpi
```
